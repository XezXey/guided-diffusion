```python
import glob, os, json, shutil

n_frame=2
params = ['shape', 'pose', 'exp', 'cam']
noise_lvl = ['1e-1', '5e-1', '1', '15e-1', '2', '3', '5']
model = "log=Masked_Face_woclip+BgNoHead+shadow_cfg=Masked_Face_woclip+BgNoHead+shadow.yaml"
sampling_dir = "/data/mint/sampling/Robustness/MPEval/"
pair_pth = '/home/mint/guided-diffusion/sample_scripts/paper_script/sample_json/paper/multipie_valset.json'


with open(pair_pth, 'r') as f:
    pair_name = json.load(f)
    

print(pair_name)
for p in params:
    for nl in noise_lvl:
        output_dir = f"/data/mint/robustness_for_evaluation/{model}/{p}/{model}_{p}_{nl}/"
        os.makedirs(output_dir, exist_ok=True)
        counter = 0
        for k, v in pair_name['pair'].items():
            src = v['src']
            dst = v['dst']
            sj_path = f"/{sampling_dir}/{model}_{p}_{nl}/ema_050000/valid/render_face/reverse_sampling/src={src}/dst={dst}/Lerp_1000/n_frames={n_frame}/"
            
            copy_fn = f'/{output_dir}/input={src}#pred={dst}.png'
            from_fn = f'{sj_path}/res_frame{n_frame-1}.png'
            shutil.copyfile(src=from_fn, dst=copy_fn)
            counter+=1
        assert counter != len(v)
```
