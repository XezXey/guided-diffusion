```python
import numpy as np
import torch as th
import torchvision
from PIL import Image
import cv2
import imageio

image_exceed = th.randint(low=0, high=300, size=(1, 3, 10, 10)) / 255.
np.save(arr=image_exceed, file='image_exceed.npy')
```

# Write Image

```python
torchvision.utils.save_image(image_exceed, fp='./image_clipped_torchvision.png')
cv2.imwrite(filename='./image_exceed_cv2.png', img=image_exceed[0].permute(1, 2, 0).numpy())
imageio.imwrite('./image_exceed_imageio.png', image_exceed[0].permute(1, 2, 0).numpy())

print(image_exceed[0].permute(1, 2, 0).shape)
image_exceed_arr = Image.fromarray(image_exceed[0].permute(1, 2, 0).numpy())
image_exceed_arr.save('./image_exceed.png')
```

# Load Image

```python
image_exceed_npy = np.load('./image_exceed.npy', allow_pickle=True)
image_clipped = np.array(Image.open('./image_clipped_torchvision.png'))
image_exceed_cv2 = np.array(Image.open('./image_exceed_cv2.png'))
image_exceed_imageio = np.array(Image.open('./image_exceed_imageio.png'))
print(image_exceed_npy)
print(image_clipped)
print(image_exceed_cv2)
print(image_exceed_imageio)
# image_clipped = Image.open('./image_clipped.png')
```
