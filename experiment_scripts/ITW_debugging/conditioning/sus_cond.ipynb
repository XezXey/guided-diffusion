```python

def examine_params(params):
    for p in params:
        aligned = f'../debugging_img/aligned_images/params/valid/ffhq-valid-{p}-anno.txt'
        original = f'../debugging_img/original_images/params/valid/ffhq-valid-{p}-anno.txt'

        aligned_dat = pd.read_csv(aligned, sep=' ', header=None)
        aligned_dict = {}
        for i in range(len(aligned_dat)):
            img_name = aligned_dat.iloc[i, 0].split('.')[0]
            aligned_dict[img_name] = aligned_dat.iloc[i, 1:].values
            
        original_dat = pd.read_csv(original, sep=' ', header=None)
        original_dict = {}
        for i in range(len(original_dat)):
            img_name = original_dat.iloc[i, 0].split('.')[0]
            original_dict[img_name] = original_dat.iloc[i, 1:].values
            
        cosim_l = []
        l1_l = []
        l2_l = []
        for k in aligned_dict.keys():
            cosim = np.dot(original_dict[k], aligned_dict[k]) / (np.linalg.norm(original_dict[k]) * np.linalg.norm(aligned_dict[k]))
            cosim_l.append(cosim)
            l2 = np.mean((original_dict[k] - aligned_dict[k])**2)
            l2_l.append(l2)
            l1 = np.mean(np.abs(original_dict[k] - aligned_dict[k]))
            l1_l.append(l1)
            # print(k, cosim)
            
        print("#"*50)
        print(p)
        print(f"Cosim of {p} : ", np.mean(cosim_l), np.std(cosim_l))
        print(f"L2 of {p} : ", np.mean(l2_l), np.std(l2_l))
        # print(f"L1 of {p} : ", np.mean(l1_l), np.std(l1_l))
        # print(l2_l)
        # print(original_dict[k])
        # print(aligned_dict[k])
        # print(f"EQ of {p} : ", np.allclose(original_dict[k], aligned_dict[k]))
            

params = ['faceemb', 'light', 'shape', 'pose', 'exp', 'cam', 'shadow']
examine_params(params)
```
