```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image

s = 128
s = 256

out = []
for i in ['./out_sd_AAx1.png', './out_sd_AAx2.png', './out_sd_AAx3.png']:
    img = Image.open(i)
    img = img.resize((s, s), Image.LANCZOS)
    out.append(img)

out = np.hstack(out)
Image.fromarray(out).save(f'out_sd_AAx123_{s}_LZ.png')

plt.imshow(out)
plt.show()


```
