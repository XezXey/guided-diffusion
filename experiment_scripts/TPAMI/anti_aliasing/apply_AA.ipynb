```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image

s = 128

out = []
for i in ['./out_sd_AAx1.png', './out_sd_AAx2.png', './out_sd_AAx3.png']:
    img = Image.open(i)
    img = np.array(img)
    img = cv2.resize(img, (s, s), interpolation=cv2.INTER_AREA)
    # img = img.resize((s, s), Image.LANCZOS)
    out.append(img)

out = np.hstack(out)
Image.fromarray(out).save(f'out_sd_AAx123_{s}_AR.png')

plt.imshow(out)
plt.show()


```
