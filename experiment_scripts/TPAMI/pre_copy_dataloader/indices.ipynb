```python
import torch
from torch.utils.data import DataLoader, TensorDataset, Dataset

# Assume you have some data and labels
data = torch.randn((1000, 64))  # Example: 1000 samples of 3x32x32 images
labels = torch.arange(1, 1001) # Example: 1000 labels, each between 0 and 9

# Create a PyTorch dataset
dataset = TensorDataset(data, labels)

# Create a DataLoader with batch_size=64 and shuffle=True
class CustomDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, index):
        return self.data[index], self.labels[index], index
    

dataloader = DataLoader(CustomDataset(data, labels), batch_size=64, shuffle=True)

# Get all indices for one epoch
all_indices = []
all_labels = []
all_true_indices = []
for batch_data, batch_labels, index in dataloader:
    batch_indices = list(range(len(all_indices), len(all_indices) + len(batch_labels)))
    all_indices.extend(batch_indices)
    all_true_indices.extend(index.tolist())
    all_labels.extend(batch_labels.tolist())
    

# 'all_indices' now contains all the indices for one epoch
print(all_indices)
print(all_labels)
print(all_true_indices)
```
